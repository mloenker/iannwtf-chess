```python
from read_games import read_games
import matplotlib.pyplot as plt
import re
games = read_games("../lichess_db_2.pgn", 10000, False)
```

```python
# numbers from 1 to 10000 in steps of 100
datapoints_x = [i for i in range(100, 10001, 100)]
datapoints_y = []

possible_moves = []
for number in datapoints_x:
    for game in games[number-100:number]:
        game = re.sub(r'\d+\.', '', game) # remove move numbers
        for move in game.split():
            if move not in possible_moves:
                possible_moves.append(move)   
    datapoints_y.append(len(possible_moves))

# plot number of possible moves vs number of games
plt.plot(datapoints_x, datapoints_y)
print(possible_moves[1000:1010]) # some sample moves
```

The number of moves increases with a logarithmic scale, creating a set vocabulary with tokenized pgn moves is thus probably unfeasible

```python
games_lengths = []
for game in games:
    game = re.sub(r'\d+\.', '', game) # remove move numbers
    games_lengths.append(len(game.split()))

print("Average game length: ", sum(games_lengths)/len(games_lengths))
```
